```python
import text2emotion as te
import pandas as pd
import numpy as np
item_list = ["iphone", "ipad", "macbook", "airpods", "apple watch"]

def get_long(comment):
    if len(comment) < 20:
        return 0
    else:
        return 1
motion_list = ['Happy', 'Angry', 'Surprise', 'Sad', 'Fear']
def get_motion(comment):
    L = []
    result = te.get_emotion(comment)
    for i in range(5):
        L.append(result[motion_list[i]])
        L[0] += 0.01
#     print(motion_list[np.argmax(L)])
    return motion_list[np.argmax(L)]
def get_positive(motion):
    if motion == 'Happy' or motion == 'Surprise':
        return 1
    else:
        return 0
for i in range(5):
    file = pd.read_csv(item_list[i]+'.csv')
    file['is_long'] = file.apply(lambda x: get_long(x.comment), axis=1)
    file['motion'] = file.apply(lambda x: get_motion(x.comment), axis=1)
    print('Done')
    file['is_positive'] = file.apply(lambda x: get_positive(x.motion), axis=1)
    print(item_list[i])
    for j in range(5):
        print(motion_list[j], ':', np.sum(file['motion']==motion_list[j]))
    print('postive:', np.sum(file['is_positive'] == 1))
    print('negetive:', np.sum(file['is_positive'] != 1))
    print('long:', np.sum(file['is_long'] == 1))
    print('notlong:', np.sum(file['is_long'] != 1))
    file.to_csv('new' + item_list[i] + '.csv', index=False)
```
